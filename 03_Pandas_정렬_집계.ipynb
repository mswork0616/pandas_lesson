## 정렬
### index명/컬럼명을 순 정렬
- #### sort_index(axis, ascending=True)
    - axis
        - index명 기준 정렬(행) : 'index' 또는 0 (기본값)
        - columnm 명 기준 정렬(열) : 'columns' 또는 1
    - ascending
        - 정렬방식
        - True(기본): 오름차순, False: 내림차순
    - inplace
        - 원본에 적용 여부
        - False(기본): 변경한 복사본 반환 
        - True : 원본을 변경

```python
fls.sort_index(axis = 1)
```

### 컬럼 값 기준 정렬
- #### sort_values(by, ascending, inplace)
    - by
        - 정렬기준 컬럼명
        - 여러 컬럼에 대해 정렬할 경우 리스트에 담아 전달
    - ascending
        - 정렬방식
        - True(기본): 오름차순, False: 내림차순
        - 여러컬럼에 대해 정렬할 경우 정렬방식도 리스트에 담아 전달
    - inplace
        - 원본에 적용 여부
        - False(기본): 변경한 복사본 반환 
        - True : 원본을 변경
    - 결측치는 방식과 상관없이 마지막에 표시

```python
# 컬럼의 값 기준 정렬 (오름차순)
fls.sort_values('ORG_AIR') # sql[select * from fls order by ORG_AIR]
```

```python
# 컬럼의 값 기준 정렬 (내림차순)
fls.sort_values('ORG_AIR', ascending=False) # sql[select * from fls order by ORG_AIR desc]
```

```python
# 여러 기준 설정 정렬
fls.sort_values(['ORG_AIR', 'DEST_AIR'])[['ORG_AIR', 'DEST_AIR']] #  sql[select * from fls order by ORG_AIR, DEST_AIR]
```

```python
# 여러 기준 정렬 순서 변경
fls.sort_values(['ORG_AIR', 'DEST_AIR'], ascending=[True, False])[['ORG_AIR', 'DEST_AIR']] # ORG_AIR는 오름차순, DEST_AIR는 내림차순
```

```python
# 특정 조건 결과를 정렬
fls[fls['MONTH'] > 6].sort_values('AIRLINE') # sql[select * from fls where MONTH > 6 order by AIRLINE] : NONTH가 6을 초과한 모든 정보를 AIRLINE 기준으로 오름차순 정렬
```

### aggregate(func, axis=0, \*args, \*\*kwargs) 또는 agg(func, axis=0, \*args, \**kwargs)
- DataFrame, Series의 메소드로 집계결과를 다양한 형태로 묶어서 볼 때 사용
    - **사용자 정의 집계메소드를 사용할 때도 편리**
- 매개변수
    - func 
        - 집계 함수 지정
            - 함수명/함수리스트 (문자열)
                - 판다스 제공 집계메소드들은 문자열로, 사용자정의 집계함수는 함수 객체로 전달
            - 딕셔너리 : {'집계할컬럼' : 집계함수, ... }
                - 컬럼별로 다른 집계를 할 경우
    - axis
        - 0 또는 'index' (기본값): 컬럼 별 집계
        - 1 또는 'columns': 행 별 집계
    - \*args, \**kwargs 
        - 함수에 전달할 매개변수
        - 집계함수는 첫번째 매개변수로 Series를 받음, 그 이외의 매개변수가 있는 경우
        
- 가변인자
    - 함수에 전달하는 값은 통계함수가 하나일 때만 적용
    - 함수에 전달하는 인자값은 사용자 정의 통계함수에서 agg를 이용하여 사용 시 적용 

```python
# Series에 적용
# 다수 통계량 : 메소드를 리스트로 묶음 (pandas에서 제공하는 메소는 이름만 명시)
# 도착지연시간의 mean(), min(), max()
fls['ARR_DELAY'].agg(['mean', 'min', 'max'])
```

```python
# DataFrame에 적용
fls.agg(['min', 'max'])
```

```python
# 컬럼 별 다른 통계를 조회할 경우
fls.agg({
    'ARR_DELAY':'mean',
    'DEP_DELAY':['mean', 'sum'],
    'AIR_TIME':['min', 'max']
})
```

# Groupby
- 특정 열을 기준으로 데이터셋을 묶음
- ~~ 별 집계를 할 때 사용
- 구문
    - DF.groupby('그룹으로묶을기준컬럼')['집계할 컬럼'].집계함수()
        - groupby에 여러개의 컬럼을 기준으로 나눌 경우 리스트에 묶어서 전달
        - 집계할 컬럼이 여러개인 경우 리스트로 묶음
    - 집계함수
        - 기술통계 함수들
        - agg()/aggregate()
            - 여러 다른 집계함수 호출 시(여러 집계를 같이 볼경우)
            - 사용자정의 집계함수 호출 시
            - 컬럼별로 다른 집계함수들을 호출할 경우

```python
# 컬럼 별 집계
g = fls.groupby('AIRLINE') # type(g) : DataFrameGroupBy
# 그룹 묶음 확인
g.groups
```

```python
# AIRLINE 별 DEP_DELAY, ARR_DELAY의 평균
fls.groupby('AIRLINE')[['DEP_DELAY', 'ARR_DELAY']].mean() # sql[select avg(DEP_DELAY), avg(ARR_DELAY) from fls group by AIRLINE]
```

```python
# AIRLINE 별 DEP_DELAY의 최대/최소, ARR_DELAY의 합계/평균
fls.groupby('AIRLINE').agg({'DEP_DELAY':['min', 'max'], 'ARR_DELAY':['sum', 'mean']}) # sql[select min(DEP_DELAY), max(ARR_DELAY), sum(DEP_DELAY), mean(ARR_DELAY) from fls group by AIRLINE]
```

## 사용자 정의 집계함수를 만들어 적용

### 사용자 정의 집계 함수 정의
- 매개변수
    1. Series 또는 DataFrame을 받을 매개변수 (필수)
    2. 필요한 값을 받을 매개변수를 선언 (선택)

### agg() 를 사용해 사용자 정의 집계 함수 호출
- DataFrame.agg(func=None, axis=0, \*args, \*\*kwargs)
    - axis : 사용자 정의 함수에 전달할 값들(Series)의 축 지정
- Series.agg(func=None, axis=0, \*args, \*\*kwargs)
    - DataFrame의 agg와 매개변수 구조를 맞추기 위해 axis 지정
        - kwargs를 이용해 매개변수 전달할 경우 axis는 생략 가능 (∵ 기본값 axis = 0)
        - keyword 인자 뒤에 position 인자는 안되서 *args로 값 전달시에는 axis를 지정해야 함
- DataFrameGroupBy.agg(func, \*args, \*\*kwargs) : 
    - axis 지정안함.
    - 사용자 함수에 Series를 group 별로 전달
- SeriesGroupBy.agg(func=None,  \*args, \*\*kwargs)  
    - axis 지정안함
    - 사용자 함수에 Series를 group 별로 전달
  
- \*args, \*\*kwargs는 사용자 정의 함수에 선언한 매개변수가 있을 경우 전달할 값을 전달  
    - 키워드 인자를 이용해 가변인자로 전달하는 것이 편리

```python
# 매개변수 없는 사용자 정의 함수 - 최대 값과 최소값의 차이를 반환


# 사용자 정의 집계 함수 생성
# 매개변수로 Series나 DataFrame을 받을 변수 선언 후 필요한 변수를 추가 선언
def max_min_diff(X): 
    # X : Series/DataFrame을 받을 변수
    # X가 수치형(int, float)이 아닌 경우 Exception 발생
    if X.dtype == 'object':
        raise TypeError('숫치형 데이터 타입만 계산 가능')
    return X.max() - X.min()
```

```python
max_min_diff(fls['ARR_DELAY'])
```

```python
# ARR_DELAY의 최소값, 최대값, 두 값의 차
# pandas 통계량 함수 : 함수 이름들 문자열로 제공
# 사용자 정의 함수 : 함수 객체
fls['ARR_DELAY'].agg(['min', 'max', max_min_diff]) # ∵ max_min_diff는 사용자 정의 함수
```

```python
# AIRLINE 별 DEP_DELAY, ARR_DELAY의 최소값, 최대값, 두 값의 차
fls.groupby('AIRLINE')[['DEP_DELAY', 'ARR_DELAY']].agg(['min', 'max', max_min_diff])
```
